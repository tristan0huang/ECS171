```python
data = pd.read_csv('/content/drive/MyDrive/ECS_171_Files/ECS171/popular_10000_movies_tmdb.csv')
```

```python
data = data.drop(columns=["title"])
data = data.drop(columns=['tagline'])
data = data.drop(columns=['overview'])
data = data.drop(columns=['original_language'])
```

```python
data = data.replace({'release_date':'-'}, {'release_date':''}, regex=True)
```

```python
data = data.dropna()
```

```python
budget_mean = math.floor(data['budget'].mean())
data = data.replace({'budget':0}, {'budget':budget_mean}, regex=True)

revenue_mean = math.floor(data['revenue'].mean())
data = data.replace({'revenue':0}, {'revenue':revenue_mean}, regex=True)
```

```python
lengths = []
for row in data['genres']:
    lengths.append(row.count(',') + 1)

data['genres'] = lengths

lengths = []
for row in data['production_companies']:
    lengths.append(row.count(',') + 1)

data['production_companies'] = lengths
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['budget'], data['revenue'], test_size=0.1, random_state=42)
```

```python
reg = LinearRegression()
regmodel = reg.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
```

```python
yhat_train = reg.predict(X_train.values.reshape(-1,1))
yhat_test = reg.predict(X_test.values.reshape(-1,1))
```

```python
print('\nMean squared error Train: %.2f' % mean_squared_error(y_train, yhat_train))
print('\nMean squared error Test: %.2f' % mean_squared_error(y_test, yhat_test))
```
